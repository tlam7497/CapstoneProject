### Checking if a text is in English

```python
gr = pd.read_csv('overall_goodreads.csv')
```

```python
from langdetect import detect, DetectorFactory
DetectorFactory.seed = 0  # Makes method deterministic (otherwise non-determininistic)
non_eng = [] # stores indexes of non-English reviews
non_detect = [] # stores indexes of reviews langdect fails to detect -> manual cleaning
for i,text in enumerate(gr.reviewText):
    try:
        if detect(text) != 'en':
            non_eng.append(i)
    except:
        non_detect.append(i)
```

```python
cleaned_gr = gr.drop(gr.index[[non_eng]])
```

```python
cleaned_gr
```

```python
# Export cleaned dataframe to csv
cleaned_gr.to_csv('cleaned_goodreads.csv',index=False)
```

```python
# Dataframe of undetected reviews
undetected_gr = gr.iloc[non_detect]
```

```python
undetected_gr
```

```python
# Export undetected reviews to csv
undetected_gr.to_csv('undetected_goodreads.csv',index=False)
```
